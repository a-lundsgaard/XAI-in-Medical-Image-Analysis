```python
import numpy as np
import matplotlib.pyplot as plt
p = 0.1

circle_radius_list = []

np.random.seed(42)
for i in range(10000):
    # Simulate integers from a uniform distribution
    circle_radius =  np.random.randint(5, 100)
    should_reduce_radius = np.random.choice([0, 1], p=[p, 1-p])
    if should_reduce_radius:
        circle_radius *= 0.5

    circle_radius_list.append(circle_radius)

# Plot a histogram
plt.hist(circle_radius_list, bins=20)
plt.xlabel('Integers')
plt.ylabel('Frequency')
plt.title('Uniform Distribution of Integers')
plt.show()
```
